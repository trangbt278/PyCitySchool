### Project Description
* Aggregate the data to and showcase obvious trends in school performance.
* Help the  school board and mayor make strategic decisions regarding future school budgets and priorities.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(pd.unique(school_data_complete['school_name']))
#Calculate the total number of students
total_students = school_data_complete['student_name'].count()
#Calculate the total budget
total_budget = school_data['budget'].sum()
#Calculate the average math score 
avrg_math = round(school_data_complete['math_score'].mean(), 2)
#Calculate the average reading score
avrg_reading = round(school_data_complete['reading_score'].mean(), 2)
#Calculate the percentage of students with a passing math score (70 or greater)
total_pass_math = len(school_data_complete[school_data_complete['math_score'] >= 70])
percent_pass_math = round((total_pass_math/total_students)*100, 2)
#Calculate the percentage of students with a passing reading score (70 or greater)
total_pass_reading = len(school_data_complete[school_data_complete['reading_score'] >= 70])
percent_pass_reading = round((total_pass_reading/total_students)*100,2)
#Calculate the percentage of students who passed math **and** reading (% Overall Passing)
total_pass_overal = len(school_data_complete[(school_data_complete['reading_score'] >= 70) 
                                             & (school_data_complete['math_score'] >= 70)])
percent_pass_overal = round((total_pass_overal/total_students)*100,2)
#Create a dataframe to hold the above results
result = {"total_schools" : [total_schools],
         "total_students" : [total_students],
         "total_budget" : [total_budget],
         "avrg_math" : [avrg_math],
         "avrg_reading": [avrg_reading],
         "percent_pass_math" : [percent_pass_math],
         "percent_pass_reading" : [percent_pass_reading],
         "percent_pass_overal" : [percent_pass_overal]}
district_summary = pd.DataFrame(data=result)
#Optional: give the displayed data cleaner formatting
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
group_by_school = school_data_complete.groupby(['school_name'])
school_data_pass_overall = school_data_complete.query('math_score >=70 & reading_score >=70')
group_by_school_pass_overall = school_data_pass_overall.groupby(['school_name'])
#school Type
school_type = group_by_school['type'].first()
#total students
school_total_students = group_by_school['student_name'].count()
#total_school_budget
school_budget = group_by_school['budget'].first()
#per student_budget
school_per_student_budget = round(school_budget / school_total_students, 2)
#average Math Score
school_avrg_math = round(group_by_school['math_score'].mean(), 2)
#average reading score
school_avrg_reading = round(group_by_school['reading_score'].mean(), 2)
#% Passing Math
school_total_pass_math = group_by_school['math_score'].apply(lambda x: x[x >=70].count())
school_percent_pass_math = round((school_total_pass_math/school_total_students)*100, 2)
#% Passing Reading
school_total_pass_reading = group_by_school['reading_score'].apply(lambda x: x[x >=70].count())
school_percent_pass_reading = round((school_total_pass_reading/school_total_students)*100, 2)
#% Overall Passing
school_total_pass_overall = group_by_school_pass_overall['student_name'].count() 
school_percent_pass_overal = round((school_total_pass_overall/school_total_students)*100,2)
#Create a dataframe to hold the above results
school_summary = pd.DataFrame ({"School Type": school_type,
                               "Total Students": school_total_students,
                               "Total School Budget": school_budget,
                               "Per Student Budget": school_per_student_budget,
                               "Average Math Score": school_avrg_math,
                               "Average Reading Score": school_avrg_reading,
                               "% Passing Math": school_percent_pass_math,
                               "% Passing Reading": school_percent_pass_reading,
                               "% Overall Passing": school_percent_pass_overal })

school_summary.reset_index(drop=True)
print(school_summary.columns)
school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
avrg_per_std_budget = round(school_summary["Per Student Budget"].mean(), 2)
max_per_std_budget = school_summary["Per Student Budget"].max()
min_per_std_budget = school_summary["Per Student Budget"].min()
middle_1 = round((avrg_per_std_budget + min_per_std_budget) / 2,2)
middle_2 = round((avrg_per_std_budget + max_per_std_budget) / 2, 2)

bins = [575, 600, 625, 650, 675]
group_names = ['<600', '600-624', '625-649', '>=650']
school_summary["Spending Range"] = pd.cut(school_summary["Per Student Budget"], bins, labels = group_names, include_lowest = True )
group_spending_range = school_summary.groupby('Spending Range')

avrg_math = round(group_spending_range['Average Math Score'].mean(), 2)
avrg_reading = round(group_spending_range['Average Reading Score'].mean(), 2)
avrg_pass_math = round(group_spending_range['% Passing Math'].mean(), 2)
avrg_pass_reading = round(group_spending_range['% Passing Reading'].mean(), 2)
avrg_pass_overall = round(group_spending_range['% Overall Passing'].mean(), 2)
score_by_school_spending = pd.DataFrame({"Average Math Score" : avrg_math,
                                        "Average Reading Score" : avrg_reading,
                                        "% Passing Math" : avrg_pass_math,
                                        "% Passing Reading" : avrg_pass_reading,
                                        "Overall Passing Rate" : avrg_pass_overall})
score_by_school_spending
```
